```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import dataset
dataset = pd.read_csv("automobiles.csv")
dataset
```

# Visualize the horse power with price


```python
sns.set_style('darkgrid')
sns.set_context('notebook')
sns.relplot(x = 'horsepower', y = "price", data = dataset, hue = 'fuel-type', col = 'num-of-doors', style = 'num-of-cylinders',\
           size = 'engine-size', sizes = (0,250), color = 'red')

# Ignoring the warning
warnings.filterwarnings("ignore", message="The figure layout has changed to tight")

plt.show()

# As per analysis in this graph we can see that we have high price car in two doors and it have eight cylinders and engine size 
# is aprroximately 160 with the 160 horspower 
```
